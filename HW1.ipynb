**(b) Advantages and disadvantages?**

(b) Parametric approach is more robust, interpretable and requires a smaller dataset for training. However, this results in inflexibilty with all its corresponding disadvantages and underfitting the data. Non-parametric approaches are more flexible, with corresponding advantages and disadvantages, for example, it leads to more accurate models when used on a bigger dataset, however, it tends to overfit to random errors.

- Is there a relationship between the predictors and the response?
- Which predictors appear to have a statistically significant relationship to the response?

`````python
# Write your answer here

import statsmodels.api as sm

X_train = auto[["cylinders","displacement","horsepower","weight","acceleration","year","origin"]]
y_true = auto['mpg']

ols_sm_2 = sm.OLS(y_true, sm.add_constant(X_train.astype(float)))
ols_sm_2_results = ols_sm_2.fit()
print(ols_sm_2_results.summary())
`````

There is a strong relationship between the predictors and the responce, as the R-squared value is close to 1. The value of 0.821 means 82.1% of variance can be explained by the model, which is a decent performance. Few predictors, such as displacement, weight, year and origin, have a big statistical significance, as evidenced by their corresponding p-values. The coefficient for year, 0.7508, means that, on average, a car older by 1 year will have 0.7508 mpg more, given that the other predictors are fixed. 

**(b) Include interactions between variables such as ````horsepower```` $\times$ ````weight````, and ````displacement```` $\times$ ````acceleration````. Do any interactions appear to be statistically significant?**

`````python
# Write your answer here
auto["horsepowerXweight"] = auto["horsepower"].astype(float)*auto["weight"]
auto["displacementXacceleration"] = auto["displacement"]*auto["acceleration"]

X_train_new = auto[["cylinders","displacement","horsepower","weight","acceleration","year","origin", "horsepowerXweight", "displacementXacceleration"]]
y_true = auto['mpg']

ols_sm_added = sm.OLS(y_true, sm.add_constant(X_train_new.astype(float)))
ols_sm_added_results = ols_sm_added.fit()
print(ols_sm_added_results.summary())
`````

As the p-values for new predictors are small, the interactions between horsepower and weight, and between displacement and accelerations are significant. Moreover, it can be seen that after introducing these predictors to the training features, the R-squared value increased.

# 10. Mini-batch gradient descent (20pt)

**In the Lab 2, we learned how to implement gradient descent and stochastic gradient descent. For this question, implement mini-batch gradient descent. Set the ````batch_size```` to 10.**

`````python
from numpy.linalg import inv

boston = pd.read_csv('data/Boston.csv') # load data

X_train = boston[["lstat"]].values # regression using all input columns
X_train = sm.add_constant(X_train)
y_train = boston['medv'].values

beta_direct = inv(X_train.T.dot(X_train)).dot(X_train.T).dot(y_train)
print(beta_direct)

beta = np.random.rand(2)
alpha = 0.001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent
n = float(len(X_train)) # Number of elements in X
batch_size = 10

######################## Write your answere here ########################
cost = []
num_batches = int(n//batch_size)
for epoch in range(epochs):
    for batch in range(num_batches):
        X_batch = X_train[batch*batch_size:batch*batch_size+batch_size]
        y_batch = y_train[batch*batch_size:batch*batch_size+batch_size]
        y_pred = np.dot(X_batch, beta)
        error = y_pred - y_batch    
        dbeta = (1/batch_size) * np.dot(X_batch.T, error)
        beta = beta - alpha * dbeta

        J = (1/batch_size) * np.dot(error.T, error)
    cost.append(J)
#########################################################################
print("Learned parameters through Mini-batch descent: {}".format(beta))
print("Directly learned parameters: {}".format(beta_direct))   
plt.plot(np.arange(epochs), cost)
plt.title("Learning curve")
plt.xlabel("Epoch")
plt.ylabel(r"J($\beta$)");
`````
